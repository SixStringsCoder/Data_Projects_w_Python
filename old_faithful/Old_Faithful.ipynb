#### Determine the best times to take videos and still shots of Yellowstone's Old Faithful based on the frequency of its eruptions

- How long you should expect to wait between eruptions?
- Is there a correlation between the length of time of the eruption and the wait time between eruptions.

```python
import csv
import matplotlib.pyplot as plt

input_file = 'old_faithful.csv'
plt.figure(figsize=(7.5, 4.25))
plt.style.use('classic')

fig = plt.figure()

with open(input_file, 'r') as old_faithful_data:
    eruptions = list(csv.reader(old_faithful_data))
    
eruptions.pop(0)

eruption_times = []
waiting_times = []

for event in range(0, len(eruptions)-1):
    eruption_times.append(float(eruptions[event][0]))
    waiting_times.append(float(eruptions[event][1]))   

plt.subplot(2, 2, 1) 
plt.boxplot(eruption_times)
plt.title('Old Faithful Eruptions')
plt.xticks([1], ['Eruptions'])
plt.xlabel('Length of Eruption (mins)')

plt.subplot(2, 2, 2) 
plt.boxplot(waiting_times)
plt.title('Waiting for Old Faithful')
plt.xticks([1], ['Waiting'])
plt.xlabel('Length of Waiting Time (mins)')

plt.subplot(2, 1, 2)
plt.scatter(eruption_times, waiting_times)
plt.title('Old Faithful Eruptions')
plt.xlabel('Length of Eruption (mins)')
plt.ylabel('Times Between ruptions (mins)')

plt.tight_layout()

fig.savefig('old_faithful_presentation.png')
plt.show()
```
