```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4], [1,4,9,16], color="green", linestyle='dashdot', label="Cars")
plt.plot([2,3,4,5], [2,3,4,5], color='#2b5b84', linestyle='dashed', label="Bikes")
plt.ylabel("Important Figures")
plt.legend()
plt.show()
```
