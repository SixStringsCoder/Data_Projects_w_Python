# Exploring Titanic Survival Rate 
### Using data based on passenger attributes of 712 passengers

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")

titanic = pd.read_csv("train.csv")
cols = ["Survived", "Pclass", "Sex", "Age", "SibSp", "Parch", "Fare", "Embarked"]
titanic = titanic[cols].dropna()
titanic.head(10)
```

## Histogram with Kernel Density Estimation which creates a smoothed line chart over the histogram.

```python
sns.distplot(titanic['Fare'])
plt.show()
```

##  Generate just the Kernel Density Esimation plot using the seaborn.kdeplot() function

```python
sns.kdeplot(titanic['Age'])
plt.show()
```

##  Add back some of the histogram by setting the 'shade' parameter to True.

```python
sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel("Age")
plt.show()
```

##  Change the default seaborn style sheet to one of these presets.
- darkgrid: Coordinate grid displayed, dark background color (__Default preset__)
- whitegrid: Coordinate grid displayed, white background color
- dark: Coordinate grid hidden, dark background color
- white: Coordinate grid hidden, white background color
- ticks: Coordinate grid hidden, white background color, ticks visible

```python
sns.set_style("white")
sns.kdeplot(titanic['Age'], shade=True)
plt.xlabel('Age')
sns.despine(left=True, bottom=True)
plt.show()
```

### Use a FacetGrid instance to generate three plots that display the age distributions for each class on the Titanic

```python
g = sns.FacetGrid(titanic, col="Pclass", size=6)
g.map(sns.kdeplot,"Age",shade=True)
g.despine(left=True, bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, col="Pclass", row="Survived", size=6)
g.map(sns.kdeplot,"Age",shade=True)
g.despine(left=True, bottom=True)
plt.show()
```

## express a third condition by generating multiple plots on the same subplot in the grid and color them differently.

```python
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex")
g.map(sns.kdeplot, "Age", shade=True).add_legend()
g.despine(left=True, bottom=True)
plt.show()
```
