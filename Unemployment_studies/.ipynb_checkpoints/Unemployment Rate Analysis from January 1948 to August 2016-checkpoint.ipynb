# Unemployment Rate as a CSV from January 1948 to August 2016

```python
import pandas as pd

unrate = pd.read_csv("unrate.csv")

unrate["DATE"] = pd.to_datetime(unrate["DATE"])
unrate.head(5)
```

```python
%matplotlib inline
```

## Look for potential 'seasonality', create a line chart of unemployment rates in 1948

```python
import matplotlib.pyplot as plt

# Display first 12 rows with DATE as y-axis and Percentage VALUE as x-axis
plt.plot(unrate["DATE"].head(12), unrate["VALUE"].head(12))

plt.title("Monthly Unemployment Trends, 1948", fontsize=20, weight='bold')
plt.xlabel("Month", fontsize=14, weight='semibold', color='brown')
plt.ylabel("Unemployment Rate", fontsize=18, weight='semibold', color='brown')

plt.xticks(rotation=90, fontsize=16, color='blue')
plt.yticks(fontsize=16, color='green')

plt.show()
```

## Multiple line charts to help compare monthly unemployment trends across time. 

```python
# Make a figure object
fig = plt.figure(figsize=(12,5))
# Make axes objects with 5 rows and 1 column
ax1 = fig.add_subplot(5, 1, 1)
ax2 = fig.add_subplot(5, 1, 2)
ax3 = fig.add_subplot(5, 1, 3)
ax4 = fig.add_subplot(5, 1, 4)
ax5 = fig.add_subplot(5, 1, 5)

# Select data from 1948
x_values_48 = unrate["DATE"][0:12]
y_values_48 = unrate["VALUE"][0:12]
# plot that data on the ax1 object
ax1.plot(x_values_48, y_values_48)

# Select data from 1949
x_values_49 = unrate["DATE"][12:24]
y_values_49 = unrate["VALUE"][12:24]
# Plot that data on the ax2 object
ax2.plot(x_values_49, y_values_49)

# Select data from 1950
x_values_50 = unrate["DATE"][24:36]
y_values_50 = unrate["VALUE"][24:36]
ax3.plot(x_values_50, y_values_50)

# Select data from 1951
x_values_51 = unrate["DATE"][36:48]
y_values_51 = unrate["VALUE"][36:48]
ax4.plot(x_values_51, y_values_51)

# Select data from 1952
x_values_52 = unrate["DATE"][48:60]
y_values_52 = unrate["VALUE"][48:60]
ax5.plot(x_values_52, y_values_52)

plt.show()
```

## Using a single plot with colored lines

```python
unrate["MONTH"] = unrate["DATE"].dt.month

fig = plt.figure(figsize=(10, 6))
plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["VALUE"], c='red')
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["VALUE"], c='blue')
plt.plot(unrate[24:36]["MONTH"], unrate[24:36]["VALUE"], c='green')
plt.plot(unrate[36:48]["MONTH"], unrate[36:48]["VALUE"], c='orange')
plt.plot(unrate[48:60]["MONTH"], unrate[48:60]["VALUE"], c='black')

plt.show()
```

## Add a legend and refactor with function to handle x-y axes values, colors, and labels

```python
fig = plt.figure(figsize=(10,6))
colors = ['red', 'blue', 'green', 'orange', 'black']
years = ['1948', '1949', '1950', '1951', '1952']

for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    subset = unrate[start_index:end_index]
    points = plt.plot(subset['MONTH'], subset['VALUE'], c=colors[i], label=years[i])
    
plt.legend(loc='upper left')  

# Add title
plt.title("Monthly Unemployment Trends, 1948-1952", fontsize=20, weight='bold')
# Add x and y axes labels
plt.xlabel("Month, Integer", fontsize=16)
plt.ylabel("Unemployment Rate, Percent", fontsize=16)

plt.show()
```
