## Visualize the gender gap across college degrees
### Generate line charts to compare across all degree categories
(export the final diagram as an image file)

```python
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degrees.head(5)
```

### Generate a line chart that visualizes the historical percentage of Biology degrees awarded to women compared to men

```python
# Line colors
cb_dark_blue = (0/255, 107/255, 164/255)
cb_orange = (255/255, 128/255, 14/255)

fig, ax = plt.subplots()

men_biology_degrees = 100 - women_degrees['Biology']
ax.plot(women_degrees['Year'], women_degrees['Biology'], c=cb_dark_blue, label='Women', linewidth=3)
ax.plot(women_degrees['Year'], men_biology_degrees, c=cb_orange, label='Men', linewidth=2)

ax.set_title("Percentage of Biology Degrees Awarded By Gender")
ax.legend(loc='upper right')

ax.tick_params(bottom=False, top=False, left=False, right=False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)
ax.spines["top"].set_visible(False)
ax.spines["bottom"].set_visible(False)

plt.show()
```

### Generate line charts for four STEM degree categories on a grid for comparison

```python
major_cats = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']
fig = plt.figure(figsize=(12, 12))


for sp in range(0,4):
    ax = fig.add_subplot(2,2,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[major_cats[sp]], c=cb_dark_blue, label='Women', linewidth=2)
    ax.plot(women_degrees['Year'], 100-women_degrees[major_cats[sp]], c=cb_orange, label='Men', linewidth=2)
    # Add your code here.
    ax.set_xlim(1968,2011)
    ax.set_ylim(0, 100)
    
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)

    ax.set_title(major_cats[sp])

# Calling pyplot.legend() here will add the legend to the last subplot that was created.
plt.legend(loc='upper right')
plt.show()
```

We can conclude that:
- **Computer Science** and **Engineering** have big gender gaps dominated by men
- **Biology** & **Math and Statistics** have small gender gaps and are much more balanced

### Line up the plots horizontally ordering them by decreasing ending gender gap

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
fig = plt.figure(figsize=(18, 3))


for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=2)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=2)
    # Add your code here.
    ax.set_xlim(1968,2011)
    ax.set_ylim(0, 100)
    
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)


    # Calling pyplot.text() here with conditionals will allow positioning of text at the left and right-most plots.
    # This reduces non-data ink (which Legends use alot of) with simple text positioned near the lines.
    if sp == 0:
        plt.text(2005, 87, "Men")
        plt.text(2002, 8, "Women")
    elif sp == 5:
        plt.text(2005, 62, "Men")
        plt.text(2001, 35, "Women")
        
plt.show()
```

### Generate line charts to compare across all degree categories

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# line colors
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

fig = plt.figure(figsize=(16, 20))

for index in range(0, 18, 3):
    ax = fig.add_subplot(6, 3, index+1)
    
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[int(index/3)]], label='Women', c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[stem_cats[int(index/3)]], label='Men', c=cb_orange, linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    
    ax.set_title(stem_cats[int(index/3)])
    
    if int(index/3) != 5:
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
        
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_yticks([0, 100])  
    ax.axhline(50, c=(181/255, 181/255, 181/255), alpha=0.3)
    
    if int(index/3) == 0:
        ax.text(2001, 87, 'Women')
        ax.text(2002, 8, 'Men')
    elif int(index/3) == 5:
        ax.text(2005, 90, 'Men')
        ax.text(2005, 5, 'Women')

        
for index in range(1, 16, 3):
    ax = fig.add_subplot(6, 3, index+1)
    
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[int((index-1)/3)]], label='Women', c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[int((index-1)/3)]], label='Men', c=cb_orange, linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    
    ax.set_title(lib_arts_cats[int((index-1)/3)])
    
    if int((index-1)/3) != 4:
        ax.tick_params(bottom=False, top=False, left=False, right=False, labelbottom=False)
        
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_yticks([0, 100])  
    ax.axhline(50, c=(181/255, 181/255, 181/255), alpha=0.3)
    
    if int((index-1)/3)  == 0:
        ax.text(2003, 78, 'Women')
        ax.text(2005, 18, 'Men')
        
        
for index in range(2, 20, 3):
    ax = fig.add_subplot(6, 3, index+1)
    
    ax.plot(women_degrees['Year'], women_degrees[other_cats[int((index-2)/3)]], label='Women', c=cb_dark_blue, linewidth=3)
    ax.plot(women_degrees['Year'], 100 - women_degrees[other_cats[int((index-2)/3)]], label='Men', c=cb_orange, linewidth=3)
    
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0, 100)
    
    ax.set_title(other_cats[int((index-2)/3)])
    
    if int((index-2)/3) != 5:
        ax.tick_params(bottom=False, top=False, right=False, left=False, labelbottom=False)
    
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_yticks([0, 100])
    ax.axhline(50, c=(181/255, 181/255, 181/255), alpha=0.3)
        
    if int((index-2)/3) == 0:
        ax.text(2002, 93, 'Women')
        ax.text(2005, 2, 'Men')
    elif int((index-2)/3) == 5:
        ax.text(2003, 65, 'Men')
        ax.text(2003, 30, 'Women')

plt.savefig("gender_degrees.png")        
plt.show()        
```
